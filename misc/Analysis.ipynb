```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("walmart.csv")
data.head()
```

```python
# Descriptive Statistics
data.describe()
```

```python
# Frequency distribution for categorical columns
cat_cols = ['Gender', 'Age', 'Occupation', 'City_Category', 'Marital_Status']

for cols in data[cat_cols]:
    data[cols].value_counts().plot(kind='bar')
    plt.show()
```

```python
# Sales by Gender
for i in data[cat_cols]:
    print(f"Sales by {i}: {data.groupby(i)['Purchase'].sum()}")
    print(f"Average Sales by {i}: {data.groupby(i)['Purchase'].mean()}")
    print("\n")
```
